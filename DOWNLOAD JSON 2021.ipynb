```python
import requests
import datetime
import json
import csv

file_url = 'https://bank.gov.ua/NBUStatService/v1/statdirectory/exchange?date='
file_name = 'UAH2021.json'

start_date = datetime.date(2021, 1, 1)
end_date = datetime.date(2021, 12, 31)
current_date = start_date
data_list = []

while current_date <= end_date:
    date_string = current_date.strftime('%Y%m%d')
    response = requests.get(file_url+date_string+'&json?')
    if response.status_code == 200:
        data = response.json()
        data_list.extend(data)
        print('+\n')
    else:
        print('!')
    current_date += datetime.timedelta(days=1)

with open(file_name, 'w', encoding='utf-8') as file:
    file.write('[\n')
    for i, item in enumerate(data_list):
        json.dump(item, file, ensure_ascii=False)
        if i != len(data_list) - 1:
            file.write(',')
        file.write('\n')
        print('write\n')
    file.write(']\n')
print("Done")
```

```python
with open('UAH2021.json', 'r', encoding='utf-8') as json_file, open('UAH2021.csv', 'w', newline='', encoding='utf-8') as csv_file:
    data = json.load(json_file)
    writer = csv.writer(csv_file)
    headers = ['r030', 'txt', 'rate', 'cc', 'exchangedate']  
    writer.writerow(headers)

    for row in data:
        values = [row['r030'], row['txt'], row['rate'], row['cc'], row['exchangedate']]
        writer.writerow(values)
        
print("Done")
```

```python
import boto3

def upload_file_to_s3(file_path, bucket_name, object_name=None):
    if object_name is None:
        object_name = file_path

    s3_client = boto3.client('s3')
    try:
        response = s3_client.upload_file(file_path, bucket_name, object_name)
        print("File uploaded successfully.")
    except Exception as e:
        print("Error uploading file:", e)

        
file_path = "UAH2021.csv"
bucket_name = "laba2bucket19"
upload_file_to_s3(file_path, bucket_name)

print("DONE")
```

```python
import boto3

def download_file_from_s3(bucket_name, object_name, file_path):
    s3_client = boto3.client('s3')
    try:
        s3_client.download_file(bucket_name, object_name, file_path)
        print("File downloaded successfully.")
    except Exception as e:
        print("Error downloading file:", e)

bucket_name = "laba2bucket19"
object_name = "UAH2021.csv"
file_path = "UAH2021.csv"
download_file_from_s3(bucket_name, object_name, file_path)

print("DONE")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('UAH2021.csv', delimiter=',')

usd_data = data[data['cc'] == 'USD']

usd_data['exchangedate'] = pd.to_datetime(usd_data['exchangedate'], format='%d.%m.%Y')

usd_data = usd_data.sort_values('exchangedate')

plt.plot(usd_data['exchangedate'], usd_data['rate'])
plt.xlabel('')
plt.ylabel('')
plt.title('   ')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
plt.savefig('USD.png')

eu_data = data[data['cc'] == 'EUR']

eu_data['exchangedate'] = pd.to_datetime(usd_data['exchangedate'], format='%d.%m.%Y')

eu_data = eu_data.sort_values('exchangedate')

plt.plot(eu_data['exchangedate'], eu_data['rate'])
plt.xlabel('')
plt.ylabel('')
plt.title('   ')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
